* User image_to_string function of pytesseract library which takes an image as an argument and returns a string. We can either directly print it or store this string in one variable.

```
image_path_in_colab='pic.JPG'
extractedInformation = pytesseract.image_to_string(Image.open(image_path_in_colab))
print(extractedInformation)
from google.colab.patches import cv2_imshow
img = cv2.imread(image_path_in_colab, cv2.IMREAD_UNCHANGED)
cv2_imshow(img)
```

```
img = cv2.imread('pic.JPG')
text = pytesseract.image_to_string(img)
print(text)
from google.colab.patches import cv2_imshow
cv2_imshow(img)
```

We can extract information on a high-resolution image by this method. For Blur, Noisy and colorful image we need to follow some image-processing steps like making image black and white, remove salt and pepper noise using lowpass filters such as averaging filters or Gaussian Filter, We can also make blur image sharpen by using Highpass filter such as Sobel filters. This Image Processing operation can also be implemented by the OpenCV library in python.
